# Plot the distribution of the normalized images

```python
import tifffile
import numpy as np
import matplotlib.pyplot as plt

hillshade = glob.glob('/workspace/data/topographical_indices_normalized/hillshade/*.tif')
hpmf = glob.glob('/workspace/data/topographical_indices_normalized/hpmf/*.tif')
slope = glob.glob('/workspace/data/topographical_indices_normalized/slope/*.tif')
stdon = glob.glob('/workspace/data/topographical_indices_normalized/stdon/*.tif')

hillshade_array = tifffile.imread(hillshade[0])
hpmf_array = tifffile.imread(hpmf[0])
slope_array = tifffile.imread(slope[0])
stdon_array = tifffile.imread(stdon[0])
plt.figure(figsize=(8,6))
plt.hist(hillshade_array[10], bins=100, alpha=0.5, label="Hillshade")
plt.hist(hpmf_array[10], bins=100, alpha=0.5, label="hpmf")
plt.hist(slope_array[10], bins=100, alpha=0.5, label="slope")
plt.hist(stdon_array[10], bins=100, alpha=0.5, label="stdon")
plt.legend(loc='upper right')
plt.show()
```

## Plot distribution of labels across the dataset

```python
from numpy import histogram
import collections
import matplotlib.pyplot as plt
import pandas as pd

labels = glob.glob('/workspace/data/split_data/labels/*.tif')

# extract number of pixels of each category from each image chip
list_dict = {0:[], 1:[], 2:[]}
for image in labels:
    unique, counts = np.unique((tifffile.imread(image)), return_counts=True)
    count_dict = dict(zip(unique, counts))    
    try:
        list_dict[0].append(count_dict[0])
    except Exception as KeyError:
        pass
    try:
        list_dict[1].append(count_dict[1])
    except Exception as KeyError:
        pass
    try:
        list_dict[2].append(count_dict[2])
    except Exception as KeyError:
        pass

# summarise the number of pixels in each category
counter = collections.Counter()
numberslist = []
for key, num in list_dict.items(): 
    numberslist.append((sum(num)))
    print(key,sum(num))

plt.rcParams['figure.figsize'] = [16, 16]

df = pd.DataFrame(numberslist)
x = ["Background", "Charcoal kilns", "Hunting pits"]
y = df[0]

container = plt.bar(x, y, color = "gray")
plt.bar_label(container)
```
